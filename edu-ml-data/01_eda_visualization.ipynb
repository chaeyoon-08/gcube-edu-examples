# 01. 탐색적 데이터 분석(EDA)과 시각화

데이터 분석의 첫 단계는 데이터를 **이해하는 것**입니다. 이 노트북에서는 pandas로 데이터를 다루고,
seaborn으로 분포와 관계를 시각화하는 **탐색적 데이터 분석(EDA)** 의 기본기를 익힙니다.

데이터는 scikit-learn 내장 **와인 데이터셋**(13개 화학 성분으로 와인 종류 3개를 구분)을 사용합니다.
내장 데이터라 별도 다운로드가 필요 없습니다.

1. 데이터 불러오기 및 구조 파악
2. 기술 통계
3. 분포 시각화
4. 변수 간 상관관계

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
sns.set_theme(style='whitegrid', font='NanumGothic')

print('pandas', pd.__version__, '| seaborn', sns.__version__)
```

## 1. 데이터 불러오기

scikit-learn 데이터셋을 pandas DataFrame으로 변환하면 다루기 편합니다.
`head()`로 앞부분을, `info()`로 자료형과 결측치 여부를 확인합니다.

```python
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df['target_name'] = df['target'].map(dict(enumerate(data.target_names)))

print('데이터 크기:', df.shape)
df.head()
```

```python
df.info()
```

## 2. 기술 통계

`describe()`로 각 변수의 평균·표준편차·사분위수를 한눈에 봅니다. 클래스별 데이터 개수도 확인합니다.

```python
display(df.describe())
print('\n클래스 분포:')
print(df['target_name'].value_counts())
```

## 3. 분포 시각화

주요 변수의 분포를 히스토그램과 박스플롯으로 살펴봅니다. 클래스별로 색을 나누면 변수가 분류에 유용한지 가늠할 수 있습니다.

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))
sns.histplot(data=df, x='alcohol', hue='target_name', kde=True, ax=axes[0])
axes[0].set_title('알코올 도수 분포 (와인 종류별)')
sns.boxplot(data=df, x='target_name', y='flavanoids', ax=axes[1])
axes[1].set_title('플라보노이드 함량 (와인 종류별)')
plt.tight_layout(); plt.show()
```

## 4. 상관관계

변수 간 상관계수를 히트맵으로 그리면, 서로 강하게 연관된 변수들을 빠르게 파악할 수 있습니다.

```python
plt.figure(figsize=(11, 9))
corr = df[data.feature_names].corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', center=0, square=True)
plt.title('변수 간 상관관계')
plt.tight_layout(); plt.show()
```

### 정리

- pandas로 데이터를 불러와 구조와 기술 통계를 확인하고, seaborn으로 분포·상관관계를 시각화했습니다.
- EDA는 모델링에 앞서 데이터의 특성과 문제점(이상치, 결측, 불균형)을 파악하는 필수 단계입니다.
