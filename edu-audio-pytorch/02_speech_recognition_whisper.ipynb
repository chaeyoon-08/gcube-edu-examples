# 02. 음성 인식 — Whisper

**Whisper** 는 OpenAI가 공개한 강력한 음성 인식(speech-to-text) 모델입니다.
다국어 음성을 텍스트로 변환하며, 단 몇 줄로 사용할 수 있습니다.

1. Whisper 모델 로드
2. 음성 → 텍스트 변환
3. 결과 확인

> 모델은 처음 실행 시 자동 다운로드됩니다. 모델 크기(tiny/base/small/medium)가 클수록 정확하지만 느리고 메모리를 더 씁니다.

```python
import whisper
import torch

print('GPU:', torch.cuda.is_available())
```

## 1. 모델 로드

교육용으로 `base` 모델을 사용합니다 (정확도와 속도의 균형). GPU가 있으면 자동으로 활용됩니다.

```python
model = whisper.load_model('base')
print('Whisper base 모델 로드 완료')
```

## 2. 음성 → 텍스트 변환

음성 파일을 텍스트로 변환합니다. 예제 음성은 librosa에 포함된 것을 사용합니다.
(실습에서는 직접 녹음한 `.wav`/`.mp3` 파일 경로를 넣으세요.)

```python
import librosa
audio_path = librosa.example('libri1')   # 예제 음성 (영어 낭독)
result = model.transcribe(audio_path)
print('인식 결과:')
print(result['text'])
```

## 3. 구간별 결과

Whisper는 전체 텍스트뿐 아니라 시간 구간별 텍스트(segments)도 제공합니다.

```python
for seg in result['segments'][:5]:
    print(f"[{seg['start']:5.1f}s - {seg['end']:5.1f}s] {seg['text'].strip()}")
```

### 정리

- Whisper로 음성을 텍스트로 변환하고, 구간별 타임스탬프까지 얻었습니다.
- 모델 크기를 바꿔 정확도와 속도를 조절할 수 있습니다.
- 자막 생성, 회의록 작성, 음성 명령 등 다양한 응용이 가능합니다.
