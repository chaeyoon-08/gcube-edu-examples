# 01. 오디오 특징 추출 — librosa

오디오를 딥러닝에 쓰려면 먼저 **신호를 시각화하고 특징을 추출**할 줄 알아야 합니다.
`librosa`로 파형(waveform), 스펙트로그램, 멜 스펙트로그램, MFCC를 다룹니다.

1. 오디오 로드 및 파형 시각화
2. 스펙트로그램 (주파수 성분의 시간 변화)
3. 멜 스펙트로그램 (사람 청각에 맞춘 주파수 척도)
4. MFCC (음색을 압축한 특징)

> 예제 오디오는 librosa에 포함되어 있으며 처음 실행 시 자동 다운로드됩니다. 실습에서는 `librosa.load('파일경로')`로 직접 불러올 수 있습니다.

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
print('librosa', librosa.__version__)
```

## 1. 오디오 로드 및 파형

오디오를 불러오면 파형(시간에 따른 진폭)과 샘플링 레이트(초당 샘플 수)를 얻습니다.

```python
y, sr = librosa.load(librosa.example('trumpet'))
print(f'샘플 수: {len(y)}, 샘플링 레이트: {sr} Hz, 길이: {len(y)/sr:.2f}초')

plt.figure(figsize=(12, 3))
librosa.display.waveshow(y, sr=sr)
plt.title('파형 (waveform)'); plt.xlabel('시간(초)')
plt.tight_layout(); plt.show()
```

## 2. 스펙트로그램

STFT(단시간 푸리에 변환)로 시간에 따른 주파수 성분을 봅니다. 데시벨 척도로 변환해 표시합니다.

```python
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
plt.figure(figsize=(12, 4))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB'); plt.title('스펙트로그램')
plt.tight_layout(); plt.show()
```

## 3. 멜 스펙트로그램

멜 척도는 사람의 청각 특성을 반영한 주파수 척도입니다. 멜 스펙트로그램은 오디오 딥러닝에서 CNN 입력으로 널리 쓰입니다.

```python
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)
S_dB = librosa.power_to_db(S, ref=np.max)

plt.figure(figsize=(12, 4))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', fmax=8000)
plt.colorbar(format='%+2.0f dB'); plt.title('멜 스펙트로그램')
plt.tight_layout(); plt.show()
print('멜 스펙트로그램 shape:', S.shape)
```

## 4. MFCC

MFCC(멜 주파수 켑스트럼 계수)는 음색을 압축한 소수의 특징으로, 음성 인식·화자 인식 등에 널리 쓰입니다.

```python
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
plt.figure(figsize=(12, 4))
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
plt.colorbar(); plt.title('MFCC (20 계수)')
plt.tight_layout(); plt.show()
print('MFCC shape:', mfccs.shape)
```

### 정리

- 파형 → 스펙트로그램 → 멜 스펙트로그램 → MFCC로 이어지는 오디오 특징 추출 과정을 익혔습니다.
- 멜 스펙트로그램과 MFCC는 오디오 딥러닝 모델의 표준 입력 특징입니다.
